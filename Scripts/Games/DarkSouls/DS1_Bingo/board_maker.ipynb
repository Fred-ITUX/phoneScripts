## Run setup

##### Ranges & challenges setup

```python
import os
import random
import itertools

lvl_range           = [25, 50]
stat_range          = [20, 35]

n_lvl_challenge     = random.randint(0, 1)
n_stat_challenge    = random.randint(0, 1)
n_coventant_task    = random.randint(0, 2)
n_npc_challenges    = random.randint(0, 2)
n_boss_weapons      = random.randint(0, 2)

challenge_bosses    = random.randint(1, 3)     #### boss with a unqiue challenge
standard_bosses     = random.randint(1, 3)     #### boss with no challenge

boardSize           = 5     #### 5x5
#### 3 x 1
#### 2 x 2          = 10
#### 1 x 3
entries             = pow(boardSize, 2)

n_current_challenges = n_stat_challenge + n_lvl_challenge + n_coventant_task + n_npc_challenges + n_boss_weapons + challenge_bosses + standard_bosses

otherC_tasks       = int(entries - n_current_challenges)   #### Adaptive number of tasks
```

#### File reading & array definitions

```python
startingClasses = ["Warrior","Knight","Wanderer","Thief","Bandit","Hunter","Sorcerer","Pyromancer","Cleric", "Deprived"]

startingGifts   = ["None","Divine Blessing","Black Firebomb","Twin Humanity","Binoculars","Pendant","Master Key","Tiny Being's Ring", "Old Witch's Ring"]

stats_list      = ["Vitality","Attunement","Endurance","Strength","Dexerity","Resistance","Intelligence","Faith"]

allBosses       = ["Moonlight Butterfly", "Pinwheel", "Taurus Demon", "Quelaag", "Bell Gargoyles", "Iron Golem", "Ceaseless Discharge", "Ornstein and Smough", "Four Kings", "Seath", "Bed of Chaos", "Nito", "Kalameet", "Artorias", "Manus", "Gwyn"]

boss_weapons    = ["Moonlight Butterfly Horn","Crystal Ring Shield","Quelaag's Furysword","Chaos Blade","Dragon Bone Fist","Golem Axe","Dragonslayer Spear","Smough's Hammer","Greatsword of Artorias","Cursed Greatsword of Artorias","Greatshield of Artorias","Darkmoon Bow","Tin Darkmoon Catalyst","Lifehunt Scythe","Manus Catalyst","Abyss Greatsword"]

boss_challenges = ["Fast Roll","Mid Roll","Fat Roll","Ninja Roll","No Hit","No Heal","One-Hand only","No Status","No Spells (any)","Spell Only (any)","Cast atleast 5 spells","Cast atleast 10 spells","No Blocking","No Consumables","Consumables and spells (any) only","No Buffs","No Rolling","Cursed","With a non-special +0 Weapon","With a non-special +5 Weapon","With a special +0 Weapon","With a special +5 Weapon","No HUD","Half HP (Dusk crown ring or curse)","Red Tearstone Ring Active","Blue Tearstone Ring Active"]

weapon_types    = ["Daggers","Straight Swords","Greatswords","Ultra Greatswords","Curved Greatswords","Curved Swords","Katanas","Piercing Swords","Axes","Great Axes","Hammers","Great Hammers","Fist & Claws","Spears","Halberds","Whip","Special weapon (twinkling or demon)","Boss weapon (forged or cut)"]

npc_list        = ["Alvina","Anastacia","Logan","Blacksmith Andre","Blacksmith Rickert","Blacksmith Vamos","Giant Blacksmith","All blacksmiths","Crestfallen Merchant","Crestfallen Warrior","Darkmoon Knightess (AL Firekeeper)","Kaathe","Fair Lady","Domhnall of Zena","Dusk of Oolacile","Eingyi","Elizabeth","Griggs of Vinheim","Hawkeye Gough","Ingward","Kingseeker Frampt","Knight Lautrec","Laurentius","Ciaran","Chester","Nico","Vince","Oswald","Patches","Petrus","Quelana","Reah","Shiva","Sieglinde","Siegmeyer","Solaire","Undead Merchant"]

npcOptions      = ["Kill", "Talk to"]

covenOptions    = ["Join" , "Betray"]

covenants_list  = ["Warrior of Sunlight","Princess Guard","Way of White","Darkwraith","Forest Hunter","Path of the Dragon","Blade of the Dark Moon","Gravelord Servant","Chaos Servant","3 Covenants"]

final_challenges = []

starting_stuff = f"Starting class:   {random.choice(startingClasses)}   with   {random.choice(startingGifts)}"
```

#### Randomizations & defs

```python
def shuffle():
        random.shuffle(boss_challenges)
        random.shuffle(weapon_types)
        random.shuffle(stats_list)
        random.shuffle(npc_list)
        random.shuffle(covenants_list)
        random.shuffle(otherC_list)
        random.shuffle(boss_weapons)

        random.shuffle(startingClasses)
        random.shuffle(startingGifts)
        random.shuffle(allBosses)
        random.shuffle(npcOptions)
        random.shuffle(covenOptions)

def otherC_task(otherC_list):
    return random.choice(otherC_list)


def stats_check(lvl_cap , stat):
    while lvl_cap < stat:
        lvl_cap += 1
        stat -= 1
        if lvl_cap > stat:
            break
    return lvl_cap , stat

def load_list(path):
    with open(path, "r") as f:
        return [line.strip() for line in f if line.strip()]


otherC_path      = f"{cPath}/other_challenges.txt"
otherC_list      = load_list(otherC_path)


for x in range(5):
    shuffle()
```

## Generating Challenges

```python
lvl_cap = 0
stat    = 1

if n_stat_challenge + n_lvl_challenge > 0:

    stat            = random.randint(stat_range[0], stat_range[1])
    lvl_cap         = random.randint(lvl_range[0], lvl_range[1])

    lvl_cap, stat = stats_check(lvl_cap, stat)

    statString      = f"{random.choice(stats_list)} ‚Üí {stat}"
    lvl_capString   = f"Level Cap {lvl_cap}"

    if n_stat_challenge > 0:
        final_challenges.append(statString)
    if n_lvl_challenge > 0:
        final_challenges.append(lvl_capString)

for x in range(n_boss_weapons):
    weapon      = random.choice(boss_weapons)
    boss_weapon = f"Forge {weapon}"
    boss_weapons.remove(weapon)
    final_challenges.append(boss_weapon)
    shuffle()


for x in range(n_coventant_task):
    covenant    = random.choice(covenants_list)
    action      = random.choice(covenOptions)
    covenants_list.remove(covenant)
    final_challenges.append(f"{action} {covenant}")
    shuffle()


for x in range(n_npc_challenges):
    action      = random.choice(npcOptions)
    npc_pg      = random.choice(npc_list)
    npc_list.remove(npc_pg)
    final_challenges.append(f"{action} {npc_pg}")
    shuffle()


for x in range(otherC_tasks):
    task        = random.choice(otherC_list)
    otherC_list.remove(task)
    final_challenges.append(task)
    shuffle()


for x in range(challenge_bosses):
    boss_challenge  = random.choice(boss_challenges)
    boss            = random.choice(allBosses)
    boss_challenges.remove(boss_challenge)
    allBosses.remove(boss)
    final_challenges.append(f"{boss} ‚Üí {boss_challenge}")
    shuffle()


for x in range(standard_bosses):
    boss            = random.choice(allBosses)
    allBosses.remove(boss)
    final_challenges.append(f"{boss}")
    shuffle()


for x in range(5):  #### Final array shuffle
    random.shuffle(final_challenges)
```

```python
try:
    import openpyxl
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "--user", "openpyxl"])
    import openpyxl

if lauching_from_gDrive == "no":
    import os
    from openpyxl import Workbook, load_workbook
    from openpyxl.styles import PatternFill, Font, Alignment, Border, Side
    from openpyxl.utils import get_column_letter
    from datetime import datetime

    #### Create unique sheet name
    tab_name = "Run_" + datetime.now().strftime("%Y-%m-%d_%H-%M")

    #### Load existing workbook ‚Äî OR delete invalid files
    if os.path.exists(output_file):
        try:
            wb = load_workbook(output_file)
        except Exception:
            print(f"‚ö†Ô∏è {output_file} is not a valid XLSX ‚Äî recreating it.")
            os.remove(output_file)
            wb = Workbook()
            ws0 = wb.active
            wb.remove(ws0)
    else:
        wb = Workbook()
        ws0 = wb.active
        wb.remove(ws0)

    #### Guarantee unique sheet name
    existing_tabs = wb.sheetnames
    if tab_name in existing_tabs:
        suffix = 2
        while f"{tab_name}_{suffix}" in existing_tabs:
            suffix += 1
        tab_name = f"{tab_name}_{suffix}"

    #### Create new sheet
    ws = wb.create_sheet(title=tab_name)

    #### Formatting
    fill = PatternFill(start_color="545454", end_color="545454", fill_type="solid")
    font = Font(color="FFFFFF", bold=False, size=15)
    alignment = Alignment(horizontal="center", vertical="center", wrap_text=True)
    border = Border(
        top=Side(style="thin", color="FFFFFF"),
        bottom=Side(style="thin", color="FFFFFF"),
        left=Side(style="thin", color="FFFFFF"),
        right=Side(style="thin", color="FFFFFF"),
    )

    #### Add starting class/gift
    final_challenges.append(starting_stuff)

    #### Write 5√ó5 grid
    idx = 0
    for r in range(1, boardSize + 1):
        for c in range(1, boardSize + 1):
            cell = ws.cell(row=r, column=c, value=final_challenges[idx])
            cell.fill = fill
            cell.font = font
            cell.alignment = alignment
            cell.border = border
            idx += 1

    #### Summary row
    summary_row = boardSize + 1
    ws.merge_cells(start_row=summary_row, start_column=1,
                   end_row=summary_row, end_column=boardSize)

    cell = ws.cell(row=summary_row, column=1, value=starting_stuff)
    cell.fill = fill
    cell.font = font
    cell.alignment = alignment
    cell.border = border

    #### Column widths
    for c in range(1, boardSize + 1):
        ws.column_dimensions[get_column_letter(c)].width = 32

    #### Row heights
    for r in range(1, summary_row + 1):
        ws.row_dimensions[r].height = 90

    #### Save (append new sheet)
    wb.save(output_file)

    print(f"Board created: {output_file}\nMy job here is done, exiting")
    quit(keep_kernel=True)
```

#### Write to the sheet

```python
#### Generate unique tab name (Run_YYYY-MM-DD_HH-MM)
base_name = "Run_" + datetime.now().strftime("%Y-%m-%d_%H-%M")
existing_titles = [ws.title for ws in spreadsheet.worksheets()]

if base_name not in existing_titles:
    tab_name = base_name
else:
    suffix = 2
    while f"{base_name}_{suffix}" in existing_titles:
        suffix += 1
    tab_name = f"{base_name}_{suffix}"

worksheet = spreadsheet.add_worksheet(title=tab_name, rows=str(boardSize), cols=str(boardSize))

#### Fill the board
rows = [final_challenges[i:i+boardSize] for i in range(0, len(final_challenges), boardSize)]

#### Resize sheet to fit the summary row
worksheet.resize(rows=len(rows) + 1, cols=boardSize)

#### Update the challenge rows
worksheet.update(rows)

#### Append class + gift merged row
summary_row = len(rows) + 1
worksheet.resize(rows=summary_row, cols=boardSize)

#### Write text in A
worksheet.update_cell(summary_row, 1, starting_stuff)

#### Merge A:lastColumn on the summary row
last_col_letter = gutils.rowcol_to_a1(1, boardSize).rstrip("1")
merge_range = f"A{summary_row}:{last_col_letter}{summary_row}"

worksheet.merge_cells(merge_range)
```

### Format

```python
cell_fmt = cellFormat(
    backgroundColor=color(0.33, 0.33, 0.33),
    textFormat=textFormat(bold=False, foregroundColor=color(1, 1, 1), fontSize=15),
    horizontalAlignment='CENTER',
    verticalAlignment='MIDDLE',
    wrapStrategy='WRAP',
    borders=Borders(
        top=Border('SOLID', color=color(1,1,1)),
        bottom=Border('SOLID', color=color(1,1,1)),
        left=Border('SOLID', color=color(1,1,1)),
        right=Border('SOLID', color=color(1,1,1)),
    )
)

#### Format the whole challenge grid
end_cell = gutils.rowcol_to_a1(len(rows), boardSize)
format_cell_ranges(worksheet, [(f"A1:{end_cell}", cell_fmt)])

#### Format summary row (Cell A_<last>)
format_cell_range(worksheet, merge_range, cell_fmt)

#### Match the format of the other cells
format_cell_range(worksheet, merge_range, cell_fmt)


#### Column widths
for c in range(1, boardSize+1):
    col_letter = gutils.rowcol_to_a1(1, c).rstrip("1")  #### "A", "B", ...
    set_column_width(worksheet, f"{col_letter}:{col_letter}", 220)

#### Row heights (include summary row)
for r in range(1, len(rows)+2):  #### +1 for summary row
    set_row_height(worksheet, f"{r}:{r}", 120)
```
